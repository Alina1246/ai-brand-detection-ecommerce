```python
def calculeaza_acuratete_model(path_csv):
    try:
        df = pd.read_csv(path_csv)
        corect = df["brand_corect"].astype(str).str.strip().str.lower()
        extras = df["brand_extras"].astype(str).str.strip().str.lower()
        return round((corect == extras).mean() * 100, 2)
    except:
        return 0.0

def calculeaza_timp_mediu(path_csv):
    try:
        df = pd.read_csv(path_csv)
        return round(df["timp_executie_secunde"].astype(float).mean(), 3)
    except:
        return 0.0


def cauta_brand_corect_fuzzy(titlu_input):
    titlu_input = titlu_input.lower().strip()
    titluri = df_ground_truth["nume_produs"].dropna().str.lower().str.strip().tolist()
    potriviri = get_close_matches(titlu_input, titluri, n=1, cutoff=0.75)
    if potriviri:
        titlu_gasit = potriviri[0]
        brand = df_ground_truth[
            df_ground_truth["nume_produs"].str.lower().str.strip() == titlu_gasit
        ]["nume_brand"].values
        if len(brand) > 0:
            return brand[0]
    return "Niciun rezultat"


def genereaza_radar_chart():
    import numpy as np
    import pandas as pd
    import plotly.graph_objects as go


    path_qwen = "/content/drive/MyDrive/Licenta1/Licenta/Rezultate/Rezultate_Qwen.csv"
    path_gemma = "/content/drive/MyDrive/Licenta1/Licenta/Rezultate/Rezultate_Gemma.csv"
    path_gemini = "/content/drive/MyDrive/Licenta1/Licenta/Rezultate/Rezultate_Gemini.csv"

    # Metrici(stocare)
    acc = {
        "Qwen": calculeaza_acuratete_model(path_qwen),
        "Gemma": calculeaza_acuratete_model(path_gemma),
        "Gemini": calculeaza_acuratete_model(path_gemini)
    }
    timp = {
        "Qwen": calculeaza_timp_mediu(path_qwen),
        "Gemma": calculeaza_timp_mediu(path_gemma),
        "Gemini": calculeaza_timp_mediu(path_gemini)
    }
    erori = {
        "Qwen": pd.read_csv(path_qwen)["numar_erori"].sum(),
        "Gemma": pd.read_csv(path_gemma)["numar_erori"].sum(),
        "Gemini": pd.read_csv(path_gemini)["numar_erori"].sum()
    }


    def normalize(values):
        min_val = min(values)
        max_val = max(values)
        return [(v - min_val) / (max_val - min_val + 1e-6) for v in values]

    models = ["Gemma", "Qwen", "Gemini"]

    acc_values = [acc[m] for m in models]
    acc_norm = normalize(acc_values)

    timp_values = [timp[m] for m in models]
    timp_scores = [max(timp_values) - t for t in timp_values]
    timp_norm = normalize(timp_scores)

    erori_values = [erori[m] for m in models]
    erori_scores = [max(erori_values) - e for e in erori_values]
    erori_norm = normalize(erori_scores)

    # Radar chart
    categorii = ["Acurate»õe", "Timp execu»õie", "Erori"]
    fig = go.Figure()

    for i, model in enumerate(models):

        r_vals = [
            acc_norm[i] if acc_norm[i] > 0 else 0.05,
            timp_norm[i] if timp_norm[i] > 0 else 0.05,
            erori_norm[i] if erori_norm[i] > 0 else 0.05
        ]

        fig.add_trace(go.Scatterpolar(
            r=r_vals,
            theta=categorii,
            fill='toself',
            name=model,
            mode="lines+markers",
            marker=dict(size=8),
            text=[
                f"Acurate»õe: {round(acc[model], 2)}%",
                f"Timp mediu execu»õie: {round(timp[model], 3)}s/produs",
                f"Erori: {erori[model]}"
            ],
            hoverinfo="text"
        ))

    fig.update_layout(
        polar=dict(radialaxis=dict(visible=True, range=[0, 1])),
        showlegend=True,
        title="Compararea modelelor LLM (Radar Chart)",
        height=550
    )

    return fig



def genereaza_tabel_rezultate_html():
    df = pd.read_csv("/content/drive/MyDrive/Licenta1/Licenta/Rezultate/Rezumat_rezultate.csv")

    html = """
    <div style='display: flex; justify-content: center; overflow-x: auto;'>
    <table style='border-collapse: collapse; font-size: 14px; text-align: center;'>
        <thead>
            <tr style='background-color: #4f46e5; color: white;'>
    """


    for col in df.columns:
        html += f"<th style='padding: 8px 16px; border: 1px solid #ccc;'>{col}</th>"
    html += "</tr></thead><tbody>"

    for _, row in df.iterrows():
        html += "<tr>"
        for val in row:
            html += f"<td style='padding: 6px 12px; border: 1px solid #ccc;'>{val}</td>"
        html += "</tr>"

    html += "</tbody></table></div>"
    return html

```
